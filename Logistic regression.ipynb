```python
import torch
import torch.utils.data as Data
from torch import nn
from torch.nn import init
import torch.optim as optim
from tqdm import tqdm
```

# 生成data

```python
num_inputs = 3
num_examples = 1000
true_w = [2, -3.4, 1]
true_b = 4.2
features = torch.randn(num_examples, num_inputs,
                       dtype=torch.float32)

y = true_w[0] * features[:, 0] + true_w[1] * features[:, 1] + true_w[2] * features[:, 2] + true_b

y_sigmoid = 1/(1+torch.exp(-y))
z = torch.distributions.bernoulli.Bernoulli(y_sigmoid)
labels = z.sample()

print(features.shape)
print(labels.shape)
batch_size = 100
# 将训练数据的特征和标签组合
dataset = Data.TensorDataset(features, labels)
# 随机读取小批量
data_iter = Data.DataLoader(dataset, batch_size, shuffle=True)
```

***重點： 即使是classification， layer仍然都是全連接的 linear 层，只需要在loss function那裡改成 crossentrophy 
实现logistic regression 有两种方法***

1： 添加一个sigmoid层然后使用BCELoss

2:  只有linear曾然后使用BCEWithLogitsLoss

# 只有linear曾然后使用BCEWithLogitsLoss 

```python
n_feature = 3
n_output = 2
batch_size = 100

class LogitReg(nn.Module):
    def __init__(self):
        super(LogitReg, self).__init__()
        self.linear = nn.Linear(3, 1)
    # forward 定义前向传播
    def forward(self, x):
        y = self.linear(x)
        return y


net = LogitReg()
print(net) # 使用print可以打印出网络的结构
for param in net.parameters():
    print(param)
    
# init parameters, loss function and optimization tools
init.normal_(net.linear.weight, mean=0, std=0.01)
init.constant_(net.linear.bias, val=0)  # 也可以直接修改bias的data: net[0].bias.data.fill_(0)

# criterion = torch.nn.BCEWithLogitsLoss()
# optimizer = optim.SGD(net.parameters(), lr=0.03)

# logistic_model = LogisticRegression()
# 定义损失函数和优化器
loss = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.9)
```

```python
num_epochs = 100
for epoch in range(num_epochs):  # 训练模型一共需要num_epochs个迭代周期
    # 在每一个迭代周期中，会使用训练数据集中所有样本一次（假设样本数能够被批量大小整除）。X
    # 和y分别是小批量样本的特征和标签
    for X, y in data_iter:
        #print(X)
        #print(y)
        y_hat = net(X)
        # now y_hat is 1000*1, we convert it into 1000
        l = loss(y_hat.view(-1), y).sum()
        #print(l)
         # 梯度清零
        if optimizer is not None:
            optimizer.zero_grad()
        elif params is not None and params[0].grad is not None:
            for param in params:
                param.grad.data.zero_()
        l.backward()
        optimizer.step()
    #print(l)

print(net.linear.weight)
print(net.linear.bias)
print(true_w, true_b)
```

# 添加一个simoid层使用BCELoss

```python
n_feature = 3
n_output = 2
batch_size = 100

class LogitReg(nn.Module):
    def __init__(self):
        super(LogitReg, self).__init__()
        self.linear = nn.Linear(3, 1)
        self.sm = nn.Sigmoid()
    # forward 定义前向传播
    def forward(self, x):
        y = self.linear(x)
        return self.sm(y)


net = LogitReg()
print(net) # 使用print可以打印出网络的结构
for param in net.parameters():
    print(param)
    
# init parameters, loss function and optimization tools
init.normal_(net.linear.weight, mean=0, std=0.01)
init.constant_(net.linear.bias, val=0)  # 也可以直接修改bias的data: net[0].bias.data.fill_(0)

# criterion = torch.nn.BCEWithLogitsLoss()
# optimizer = optim.SGD(net.parameters(), lr=0.03)

# logistic_model = LogisticRegression()
# 定义损失函数和优化器
loss = nn.BCELoss()
optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.9)
```

```python
num_epochs = 20
for epoch in range(num_epochs):  # 训练模型一共需要num_epochs个迭代周期
    # 在每一个迭代周期中，会使用训练数据集中所有样本一次（假设样本数能够被批量大小整除）。X
    # 和y分别是小批量样本的特征和标签
    for X, y in data_iter:
        #print(X)
        #print(y)
        y_hat = net(X)
        # now y_hat is 1000*1, we convert it into 1000
        l = loss(y_hat.view(-1), y).sum()
        #print(l)
         # 梯度清零
        if optimizer is not None:
            optimizer.zero_grad()
        elif params is not None and params[0].grad is not None:
            for param in params:
                param.grad.data.zero_()
        l.backward()
        optimizer.step()
    print(l)

print(net.linear.weight)
print(net.linear.bias)
print(true_w, true_b)
```
